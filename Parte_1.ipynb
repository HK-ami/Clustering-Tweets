```python
# Removendo menções
import re
def remove_mention(tweet):
#     return re.sub(r'@[A-Za-z0-9]+','',tweet)
    return re.sub("@([a-zA-Z0-9_]{1,50})",'',tweet)
df['Tweet'] = df['Tweet'].apply(remove_mention)
df['Tweet'].replace({"#": ""}, regex=True) # acho que não ta funcnionando
# removing all characteres including numbers: re.sub("[^a-zA-Z]", " ", df.Tweet[])
```

```python
# Removendo URL
import re
def remove_url(tweet):
#     return re.sub(r'@[A-Za-z0-9]+','',tweet)
    return re.sub("https?://[A-Za-z0-9./]+",'', tweet)
df['Tweet'] = df['Tweet'].apply(remove_url)
```

```python
# Deixando apenas letras
def only_letters(tweet):
    return re.sub("[^A-Za-záàâãéèêíïóôõöúçñÁÀÂÃÉÈÍÏÓÔÕÖÚÇÑ]", " ", tweet)

df['Tweet'] = df['Tweet'].apply(only_letters)
df
```

```python
for item in more_words_list:
    if item not in stopwords:
        stopwords.append(item)
len(stopwords)
```

```python
# Adicionando mais palavras que desejamos remover da análise.
list = ['retweet', 'rts', 'rt', 'tá', 'ta', 'to', 'tô', 'tamo', 'vai', 'vamos', 'vamo', 'n', 
        'tbm', 'mto', 'sla', 'plmds', 'mds', 'q', 'pq', 'ne', 'né', 'p', 'pra', 'pro', 'pandemia']
for item in list:
    stopwords.append(item)
len(stopwords)
```

```python
df['text'] = df['Tweet'].str.lower()\
          .apply(lambda x: [i for i in x.split() if not i in stopwords])
df = pd.DataFrame(df, columns=['text'])

df['text'] = df['text'].str.join(" ")
df
```

```python
# Removendo duuplicatas
df.drop_duplicates(inplace=True)
df
```

```python
# Rearranjando os índices
df = df.reset_index(drop=True)
df
```

```python
clean_df = pd.DataFrame(df, columns=['text'])
clean_df.head()
clean_df.to_csv('clean_tweet.csv', encoding='utf-8')
```
